## 画像データの取り扱い  
  メタデータにて，画像データのファイル名を管理。
  実際のファイルはzipファイル内にて管理する。
+ データ処理プロセス
 1. メタデータを確認
 2. 分析手法に基づき、必要なファイル名を抽出
 3. 画像データを読み込み

```python
# Prepare meta data
import pandas as pd
df_meta = pd.DataFrame({'tar_NE':[1000,1000,1000], 'tar_GAIRCYL':[0.10,0.15,0.20],\
                        'num':[1,1,2], 'file_name':['{}.jpeg'.format(i+1) for i in range(3)]})
```

```python
df_meta
```

```python
from PIL import Image
import zipfile
import io
```

```python
def get_img(zipname, filename):
    with  zipfile.ZipFile(zipname) as zip:
            img = Image.open( io.BytesIO(zip.read(filename)) )
    return img
```

```python
zipname = 'zip.zip'
filepath = 'zip/'+df_meta[(df_meta.tar_NE == 1000)&(df_meta.tar_GAIRCYL == 0.10)].file_name[0]
```

```python
img = get_img(zipname, filepath)
```

```python
from PIL import Image
with  zipfile.ZipFile('zip.zip') as zip:
    for info in zip.infolist():
        if info.is_dir(): continue
        if info.filename in df_meta.file_name.to_list():
            img = Image.open( io.BytesIO(zip.read(info.filename)) )
```
